```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = {
    'Ripe': [1, 1, 1, 1, 0, 0],
    'Wormy': [0, 0, 1, 0, 0, 1],
    'Washed': [1, 0, 1, 0, 1, 1],
    'Eat': ['Yes', 'No', 'No', 'No', 'No', 'No']
}

df = pd.DataFrame(data)
```

```python
X = df[['Ripe', 'Wormy', 'Washed']]
y = df["Eat"]
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(X, y)
```

```python
ripe = int(input("Is the fruit ripe? (1 for Yes, 0 for No): "))
wormy = int(input("Is the fruit wormy? (1 for Yes, 0 for No): "))
washed = int(input("Is the fruit washed? (1 for Yes, 0 for No): "))
features = np.array([[ripe, wormy, washed]])
prediction = model.predict(features)
print(f"The prediction is: {prediction[0]}")
# Visualizing the decision tree
from sklearn import tree
plt.figure(figsize=(12, 8))
tree.plot_tree(model, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree for Fruit Eating Decision")
plt.show()
```
