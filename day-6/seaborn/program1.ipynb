```python
import seaborn as sns
sns.set_theme(style="whitegrid")

fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind = "line",
    x = "timepoint", y = "signal", col = "region",
    hue = "event", style = "event",
)
```

```python
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

#scatterplot
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# Lineplot: x axis = day, y axis = total_bill, hue = sex (male vs female)
sns.lineplot(x='day', y='total_bill', hue='sex', data=tips, marker='o')
plt.title("Total Bill by Day and Gender")
plt.ylabel("Total Bill")
plt.xlabel("Day")
plt.show()
```

```python
#plot a violin plot of total_bill by day
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.ylabel("Total Bill")
plt.xlabel("Day")
plt.show()
```

```python
# heatmap of total_bill by day and time
pivot_table = tips.pivot_table(values='total_bill', index='day', columns='time', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Heatmap of Average Total Bill by Day and Time")
plt.xlabel("Time")
plt.ylabel("Day")
plt.show()
```

```python
# correlation heatmap of tips dataset

numeric_tips = tips.select_dtypes(include='number')
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```
