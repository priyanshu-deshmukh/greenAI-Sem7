```python
import numpy as np
import pandas as pd
energy_data = pd.read_csv('energy_consumption.csv')
```

```python
# Display the number of null values in each column
null_values = energy_data.isna().sum()
print("Number of null values in each column:")
print(null_values)

# Replace null values with the mean of only numeric columns
numeric_cols = energy_data.select_dtypes(include='number').columns
energy_data[numeric_cols] = energy_data[numeric_cols].fillna(energy_data[numeric_cols].mean())

# Display the updated DataFrame
print("\nData after replacing null values with the mean:")
print(energy_data.head())
```
